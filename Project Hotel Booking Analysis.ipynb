### Variables Description

hotel: The type of hotel (e.g., Resort Hotel or City Hotel).

is_canceled: A binary variable indicating whether the booking was canceled (1) or not (0).

lead_time: The number of days between the booking date and the arrival date.

arrival_date_year: The year of the arrival date.

arrival_date_month: The month of the arrival date.

arrival_date_week_number: The week number of the arrival date.

arrival_date_day_of_month: The day of the month of the arrival date.

stays_in_weekend_nights: The number of weekend nights (Saturday and Sunday) the guest stayed.

stays_in_week_nights: The number of weekday nights the guest stayed.

adults: The number of adults included in the booking.

children: The number of children included in the booking.

babies: The number of babies included in the booking.

meal: The type of meal booked (e.g., Bed & Breakfast, Full Board).

country: The country of origin of the guest.

market_segment: The market segment designation (e.g., Corporate, Online TA).

distribution_channel: The distribution channel used to make the booking (e.g., Travel Agents, Direct).

is_repeated_guest: A binary variable indicating whether the guest is a repeated guest (1) or not (0).

previous_cancellations: The number of previous bookings canceled by the guest.

previous_bookings_not_canceled: The number of previous bookings not canceled by the guest.

reserved_room_type: The type of room reserved by the guest.

assigned_room_type: The type of room assigned to the guest at check-in.

booking_changes: The number of changes made to the booking.

deposit_type: The type of deposit made for the booking.

agent: The ID of the travel agency that made the booking.

company: The ID of the company or entity if the booking was made on behalf of the guest.

days_in_waiting_list: The number of days the booking was in the waiting list before being confirmed.

customer_type: The type of booking customer (e.g., Contract, Transient).

adr: The average daily rate (price per room) of the booking.

required_car_parking_spaces: The number of car parking spaces requested by the guest.

total_of_special_requests: The total number of special requests made by the guest.

reservation_status: The status of the reservation (e.g., Check-Out, Canceled).

reservation_status_date: The date on which the reservation status was last updated.

#### Chart - 1

````python
import pandas as pd
import matplotlib.pyplot as plt

# Create a new column for the date
df['date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'].astype(str) + '-01')

# Group the data by year and month and count the number of bookings and cancellations
booking_cancellation_trend = df.groupby(['date', 'reservation_status'])['hotel'].count().unstack(fill_value=0)

# Calculate the total number of bookings (excluding cancellations)
booking_cancellation_trend['Total Bookings'] = booking_cancellation_trend['Check-Out']

# Calculate the total number of cancellations
booking_cancellation_trend['Cancellations'] = booking_cancellation_trend['Canceled'] + booking_cancellation_trend['No-Show']

# Plot the line chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the trends
booking_cancellation_trend['Total Bookings'].plot(kind='line', marker='o', label='Total Bookings', ax=ax)
booking_cancellation_trend['Cancellations'].plot(kind='line', marker='o', label='Cancellations', ax=ax)

# Set labels and title
ax.set_xlabel('Month/Year')
ax.set_ylabel('Number of Bookings/Cancellations')
ax.set_title('Booking and Cancellation Trend over Time')
plt.legend()

# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

The line chart showing the Booking Cancellation Trend over Time is a valuable visualization tool to gain insights into the historical patterns of bookings and cancellations, which can help in making data-driven decisions, optimizing business strategies, and improving overall performance in the hotel industry.

##### 2. What is/are the insight(s) found from the chart?

Seasonal Patterns: The line chart reveals seasonal patterns in both bookings and cancellations. For example, there could be higher booking activity during certain months or seasons, which might correspond to increased cancellations during busy periods.

Peak Booking Periods: Peaks in the total bookings line indicates periods of high demand, which may result in a higher number of cancellations due to overbooking or changes in travel plans.

Cancellation Trends: By comparing the cancellation line with the total bookings line, it's possible to identify periods of increased cancellations relative to overall booking trends. This insight could highlight specific factors influencing cancellations, such as external events or policy changes.

#### Chart - 2

````python
# Chart - 2 visualization code
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by reserved room type and calculate the mean ADR for each type
adr_by_room_type = df.groupby('reserved_room_type')['adr'].mean().reset_index()

# Sort the data by ADR in descending order
adr_by_room_type = adr_by_room_type.sort_values(by='adr', ascending=False)

# Plot the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(adr_by_room_type['reserved_room_type'], adr_by_room_type['adr'], color='skyblue')

# Set labels and title
ax.set_xlabel('Reserved Room Type')
ax.set_ylabel('Average Daily Rate (ADR)')
ax.set_title('Average Daily Rate (ADR) by Reserved Room Type')

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
````

A bar chart will clearly show which room type has the highest or lowest ADR, allowing viewers to quickly identify trends and differences.

#### Chart - 3

````python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the revenue for each booking (ADR multiplied by the number of nights)
df['revenue'] = df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])

# Group the data by month and year and calculate the total revenue for each period
revenue_by_time = df.groupby([df['arrival_date_year'], df['arrival_date_month']])['revenue'].sum().reset_index()

# Convert the month and year columns to a single datetime column for proper sorting
revenue_by_time['date'] = pd.to_datetime(revenue_by_time['arrival_date_month'] + ' ' + revenue_by_time['arrival_date_year'].astype(str), format='%B %Y')

# Sort the data by the datetime column
revenue_by_time = revenue_by_time.sort_values(by='date')

# Plot the line chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(revenue_by_time['date'], revenue_by_time['revenue'], marker='o', color='skyblue')

# Set labels and title
ax.set_xlabel('Month/Year')
ax.set_ylabel('Total Revenue')
ax.set_title('Total Revenue Over Time')

# Format x-axis date labels to display only month and year
date_format = '%b %Y'

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Chart - 4 visualization code
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by country and count the number of bookings for each country
bookings_by_country = df['country'].value_counts()

# Keep only the top N countries with the most bookings
top_n_countries = 10
if len(bookings_by_country) > top_n_countries:
    other_countries_count = bookings_by_country[top_n_countries:].sum()
    bookings_by_country = bookings_by_country[:top_n_countries]
    bookings_by_country['Other'] = other_countries_count

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(bookings_by_country, labels=bookings_by_country.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Set title
ax.set_title('Booking Distribution by Country')

# Show the plot
plt.show()
````
